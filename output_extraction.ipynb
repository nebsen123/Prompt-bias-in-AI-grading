```python
import json
import regex
import csv
with open("outputtest.json", "r") as f:
    output = json.load(f)

with open("HPC run/data.json", "r") as f:
    data = json.load(f)
print(len(data)/10)
```

```python
def mistakes(answer_key_id):
    correct_list = "000101101001000"
    True_mistakes = 0
    for i in range(len(correct_list)):
        if int(correct_list[i])!=int(answer_key_id[i]):
            True_mistakes += 1
    return True_mistakes

prompt_types = ["very_pessimistic","pessimistic","neutral","confident","very_confident"]
count = 0
with open("ai_grading.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["answer_key_id","True_error","prompt_type","ai_error"])
for i in data[:1]:
    answer_key_id = i[:16]
    True_error = mistakes(answer_key_id)
    prompt_type = prompt_types[count%5]
    ai_error = 15-int(regex.findall(r"frac{\d}\{15\}",output[count])[0][5])
    with open("ai_grading.csv", "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([answer_key_id,True_error,prompt_type,ai_error])
    count += 1

        


```
